## Loading Data

```python
train_dataset = pd.read_csv("archive1/Training.csv").iloc[:, 2:135]
test_dataset = pd.read_csv("archive/Testing.csv").iloc[:, 0:133]
```

```python
train_dataset
```

```python
temp = train_dataset
for index, x in temp.iterrows():
  
  prefer_list = [ 'Chicken pox','Common Cold', 'Covid-19','Diabetes ','Drug Reaction','Hypertension ', 'Monkey Pox','Migraine','Tuberculosis', 'Pneumonia']
  if x[-1] in prefer_list:
    train_dataset = train_dataset.append(pd.Series(x), ignore_index=True)

```

## Splitting Data into X and Y

```python
X_train = train_dataset.iloc[:, 0:132]
y_train = pd.DataFrame(train_dataset.iloc[:, 132].values.reshape((6360, 1)))

X_test = test_dataset.iloc[:, 0:132]
y_test = pd.DataFrame(test_dataset.iloc[:, 132].values.reshape((42, 1)))
```

## Attempting to Increase Data

```python
import random
for new_turn in range(1):
  for x in train_dataset.values:
    temp = []
    for y in x[0:132]:

      prob = random.random()
      if y == 0:
        temp.append(0)
      elif type(y) == str:
        pass
      else:
        if prob > 0.6:
          temp.append(0)
        else:
          temp.append(1)
    X_train = np.append(X_train, [temp], axis = 0)
    
    y_train = np.append(y_train, np.array([[x[-1]]]), axis = 0)



  for x in test_dataset.values:
    temp = []
    for y in x[0:132]:

      prob = random.random()
      if y == 0:
        temp.append(0)
      elif type(y) == str:
        pass
      else:
        if prob > 0.6:
          temp.append(0)
        else:
          temp.append(1)
    X_test = np.append(X_test, [temp], axis = 0)
    
    y_test = np.append(y_test, np.array([[x[-1]]]), axis = 0)
print("done")
```

## Encoding Categorical Data

```python
y_train = pd.get_dummies(pd.DataFrame(y_train))
y_test = pd.get_dummies(pd.DataFrame(y_test)[0])

```

```python
diagnosis = list(y_train.columns)
diagnosis
```

# Building a Model

```python
from tensorflow.keras.layers import LeakyReLU

model = tf.keras.models.Sequential()

model.add(Input((132,)))

model.add(Dense(units = 60, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(units = 120, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(units = 240, activation="relu"))

model.add(Dense(units = 43, activation = "softmax"))
```

```python
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer="adam", metrics=["accuracy"])
```

```python
print(y_train.values.shape)
print(X_train.shape)
```

```python
model.fit(X_train, y_train.values, epochs=10)
```

```python
model.evaluate(X_test, y_test.values)
```
